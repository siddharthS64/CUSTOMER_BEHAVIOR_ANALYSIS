```python
import pandas as pd
```

```python
# importing the CSV file 
df=pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
#  summary statistics 
df.describe(include='all')
```

```python
#  checking no of nulls 
df.isna().sum()
```

```python
# filling all the null values with corresponding mean of the category
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isna().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
labels=['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df["age_group"]=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#  Creating column purcahse frequency days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30, 
    'Every 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
# checking the redundency of the values
(df['discount_applied']==df['promo_code_used']).all()
```

```python
# droping redundant column 
df.drop('promo_code_used',axis=1,inplace=True)
```

```python
df.head(10)
```

```python
# Loading data to MySQL server
import sqlalchemy as sal
engine = sal.create_engine('mysql+pymysql://root:password@localhost:3306/customer_behavior')
conn = engine.connect()
df.to_sql('customer', con=conn, if_exists='append', index=False)
conn.close()
print(f"Data successfully loaded into table 'customer' in database 'cutomer_behavior'.")
```
